### Complex Model - Round Two with Stepwise Function

```python
def stepwise_selection(X, y, threshold_in = 0.01, threshold_out = 0.05, verbose = True):
    # The function is checking for p-values (whether features are statistically significant) - lower is better
    included = [] # this is going to be the list of features we keep
    while True:
        changed = False
        # forward step
        excluded = list(set(X.columns) - set(included))
        new_pval = pd.Series(index = excluded, dtype = 'float64')
        for new_column in excluded:
            model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included + [new_column]]))).fit()
            new_pval[new_column] = model.pvalues[new_column]
        best_pval = new_pval.min()
        # we add the feature with the lowest (best) p-value under the threshold to our 'included' list
        if best_pval < threshold_in:
            best_feature = new_pval.idxmin()
            included.append(best_feature)
            changed = True
            if verbose:
                print('Add  {:30} with p-value {:.6}'.format(best_feature, best_pval)) # specifying the verbose text


        # backward step: removing features if new features added to the list make them statistically insignificant
        model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included]))).fit()

        # use all coefs except intercept
        pvalues = model.pvalues.iloc[1:]
        worst_pval = pvalues.max() # null if pvalues is empty
        # if the p-value exceeds the upper threshold, the feature will be dropped from the 'included' list
        if worst_pval > threshold_out:
            changed = True
            worst_feature = pvalues.idxmax()
            included.remove(worst_feature)
            if verbose:
                print('Drop {:30} with p-value {:.6}'.format(worst_feature, worst_pval))
        if not changed:
            break
    return included
```

```python
# Call the stepwise_selection function with X_train and y_train as parameters and store the result in step_result
step_result = stepwise_selection(X_train_scaled, y_train)

# Print the step_result
print(step_result)
```

```python
#Declare our new feature columns according to the stepwise function
fe_cols_stepped = ['Under_five_deaths', 'const', 'Adult_mortality', 
                   'Economy_status_Developed', 'Reg_Central America and Caribbean', 
                   'GDP_per_capita', 'Reg_South America', 'Reg_Oceania', 'Infant_deaths', 
                   'Reg_European Union', 'Schooling', 'BMI', 'Incidents_HIV', 'Hepatitis_B']
```

```python
# Select the new feautures columns for the new model
X_train_stepped = X_train_scaled[fe_cols_stepped]
```

```python
# Create an OLS (Ordinary Least Squares) linear regression model using the training data
lin_reg = sm.OLS(y_train, X_train_stepped)

# Fit the linear regression model to the training data
results2 = lin_reg.fit()

# Display a summary of the regression results
results2.summary()
```

```python
# Predict the values using the trained model
y_pred_stepped = results2.predict(X_train_stepped)

# Calculate the root mean squared error (RMSE) between the actual and predicted values
rmse = statsmodels.tools.eval_measures.rmse(y_train, y_pred_stepped)

# Print the RMSE
print(rmse)
```

```python
# Read the CSV file into a pandas DataFrame
df = pd.read_csv('Life Expectancy Data.csv')
```

```python
# Create a DataFrame of scaled training data using the MinMaxScaler
X_train_scaled = pd.DataFrame(minmax.transform(X_train), columns = final_cols)
X_train_scaled.head()
```
